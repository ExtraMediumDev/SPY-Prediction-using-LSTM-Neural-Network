```python
import numpy as np
import time as tm
import datetime as dt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
# Window size or the sequence length, 7 (1 week)
N_STEPS = 7

# Lookup steps, 1 is the next day, 3 = after tomorrow
LOOKUP_STEPS = [1, 2, 3]
```

N_STEPS means the number of days in our window, we are using a depth of 7 for our neural network

LOOKUP_STEPS is the number of days we are trying to predict, in this case, we are trying to predict 3

```python
# remove columns which our neural network will not use
df = df.drop(['open', 'high', 'low', 'volume'], axis=1)

display(df)
```

```python
df['timestamp'] = df.index

# If 'timestamp' is not in the correct datetime format, convert it.
df['timestamp'] = pd.to_datetime(df['timestamp'])
display(df)
```

```python
plt.style.use(style='ggplot')
plt.figure(figsize=(16,10))
plt.plot(df['close'][-200:])
plt.xlabel("days")
plt.ylabel("price")
plt.legend([f'Actual price for {"SPY"}'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['close'] = scaler.fit_transform(np.expand_dims(df['close'].values, axis=1))

#remove weekend NaN data
df = df.dropna(subset=['close'])

display(df)
```

```python
from collections import deque
def PrepareData(days):
  init_df = df.copy()
  init_df['future'] = init_df['close'].shift(-days)
  last_sequence = np.array(init_df[['close']].tail(days))
  init_df.dropna(inplace=True)
  sequence_data = []
  sequences = deque(maxlen=N_STEPS)

  for entry, target in zip(init_df[['close'] + ['timestamp']].values, init_df['future'].values):
      sequences.append(entry)
      if len(sequences) == N_STEPS:
          sequence_data.append([np.array(sequences), target])

  last_sequence = list([s[:len(['close'])] for s in sequences]) + list(last_sequence)
  last_sequence = np.array(last_sequence).astype(np.float32)

  # construct the X's and Y's
  X, Y = [], []
  for seq, target in sequence_data:
      X.append(seq)
      Y.append(target)

  # convert to numpy arrays
  X = np.array(X)
  Y = np.array(Y)

  return init_df, last_sequence, X, Y
```

```python
def GetTrainedModel(x_train, y_train):
  model = Sequential()
  model.add(LSTM(60, return_sequences=True, input_shape=(N_STEPS, len(['close']))))
  model.add(Dropout(0.3))
  model.add(LSTM(120, return_sequences=False))
  model.add(Dropout(0.3))
  model.add(Dense(20))
  model.add(Dense(1))

  BATCH_SIZE = 8
  EPOCHS = 80

  model.compile(loss='mean_squared_error', optimizer='adam')

  model.fit(x_train, y_train,
            batch_size=BATCH_SIZE,
            epochs=EPOCHS,
            verbose=1)

  model.summary()

  return model
```

Training our model

```python
predictions = []

for step in LOOKUP_STEPS:
  df, last_sequence, x_train, y_train = PrepareData(step)
  x_train = x_train[:, :, :len(['close'])].astype(np.float32)

  model = GetTrainedModel(x_train, y_train)

  last_sequence = last_sequence[-N_STEPS:]
  last_sequence = np.expand_dims(last_sequence, axis=0)
  prediction = model.predict(last_sequence)
  predicted_price = scaler.inverse_transform(prediction)[0][0]

  predictions.append(round(float(predicted_price), 2))
```
